Below are relevant functions in R for a normal RV $X∼N(\mu,\sigma^2)$:

dnorm(x, mean=mu, sd=sigma): find PDF $f_X(x)$

pnorm(x, mean=mu, sd=sigma): find CDF $F_X(x)=Pr(X\leq x)$

qnorm(p, mean=mu, sd=sigma): find quantile c such that $F_X(c)=Pr(X\leq c)=p$

rnorm(n, mean=mu, sd=sigma): generate n independent values of $X$

```r
# Create data that is normally distributed with mean 0
# and standard deviation 1
x <- seq(-5, 5, by = 0.1)
y <- dnorm(x, mean = 0.0, sd = 1.0)
# Plot the probability density function (PDF)
plot(x,y, main = "Normal Distribution PDF", col = "blue", type = "o")
```

```r
# Plot the cumulative distribution function (CDF)
y <- pnorm(x, mean = 0.0, sd = 1.0)
plot(x,y, main = "Normal Distribution CDF", col = "blue", type = "o")
```

```r
# Plot the quantiles
x <- seq(0, 1, by = 0.02)
y <- qnorm(x, mean = 0.0, sd = 1.0)
plot(x,y, main = "Normal Distribution Quantiles", col = "blue", type = "o")
```

68-95-99.7 rule:

Here we present a useful rule of thumb for the probability of falling within 1, 2, and 3 standard deviations of the mean in the normal distribution.

this will be useful in a wide range of practical settings, especially when trying to make a quick estimate without a calculator or normal distribution
table.

$\begin{align}
\Pr(\mu-\sigma \leq X\leq \mu+\sigma)&=\Pr(|Z|\leq 1) \approx 0.68; \\
\Pr(\mu-2\sigma \leq X\leq \mu+2\sigma)&=\Pr(|Z|\leq 2) \approx 0.95; \\
\Pr(\mu-3\sigma \leq X\leq \mu+3\sigma)&=\Pr(|Z|\leq 3) \approx 0.997.
\end{align}$
